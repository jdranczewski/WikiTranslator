```python
import pandas as pd
from math import log10, floor
```

```python
langlist = pd.read_csv("langlist.csv")
```

```python
langlist.head()
```

```python
langlist.iloc[langlist["Language (local)"].str.lower().argsort()].head()
```

```python
text = ""
for index, row in langlist.iloc[langlist["Language (local)"].str.lower().argsort()].iterrows():
    option = '<option value="{}">{} ({})</option>'.format(row["Wiki"], row["Language (local)"], row["Language"])
    print(option)
    text += option
```

```python
text
```

```python
def round_sig_fig(x, n=2):
    if not x: return x
    return round(x, -int(floor(log10(abs(x))))+n-1)
```

```python
text = "["
for index, row in langlist.iloc[langlist["Language (local)"].str.lower().argsort()].iterrows():
    option = '["{}","{}","{}",{}]'.format(row["Wiki"], row["Language (local)"], row["Language"], round_sig_fig(int(row["Articles"].replace(',', '')),2))
    print(option)
    text += option + ","
text += "]"
```
