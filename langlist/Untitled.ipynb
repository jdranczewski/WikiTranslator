```python
import pandas as pd
```

```python
langlist = pd.read_csv("langlist.csv")
```

```python
langlist.head()
```

```python
langlist.iloc[langlist["Language (local)"].str.lower().argsort()].head()
```

```python
text = ""
for index, row in langlist.iloc[langlist["Language (local)"].str.lower().argsort()].iterrows():
    option = '<option value="{}">{} ({})</option>'.format(row["Wiki"], row["Language (local)"], row["Language"])
    print(option)
    text += option
```

```python
text
```

```python
text = "["
for index, row in langlist.iloc[langlist["Language (local)"].str.lower().argsort()].iterrows():
    option = '["{}","{}","{}",{}]'.format(row["Wiki"], row["Language (local)"], row["Language"], int(row["Articles"].replace(',', '')))
    print(option)
    text += option + ","
text += "]"
```
